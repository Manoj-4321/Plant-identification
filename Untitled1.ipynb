```python
import pandas as pd
import matplotlib.pyplot as plt

def calculate_cagr(initial_value, end_value, num_years):
    return ((end_value / initial_value) ** (1 / num_years)) - 1

def calculate_yoy_growth(initial_value, cagr, year):
    return (initial_value * (1 + cagr) ** year) / (initial_value * (1 + cagr) ** (year - 1)) - 1

def calculate_yearly_values_with_factors(initial_value, end_value, start_year, end_year, factors_dict):
    num_years = end_year - start_year + 1
    cagr = calculate_cagr(initial_value, end_value, num_years)

    years = []
    market_values = []
    yoy_growth_values = []

    for year in range(start_year, end_year + 1):
        year_value = initial_value * (1 + cagr) ** (year - start_year)

        # Apply custom factors for specific years
        if year in factors_dict:
            year_value *= factors_dict[year]

        yoy_growth = calculate_yoy_growth(initial_value, cagr, year - start_year)

        years.append(year)
        market_values.append(year_value)
        yoy_growth_values.append(yoy_growth)

    df = pd.DataFrame({"Year": years, "Market Value (Million USD)": market_values, "YoY Growth": yoy_growth_values})
    return df

def plot_graphs(df):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(df["Year"], df["Market Value (Million USD)"], marker='o', linestyle='-')
    plt.title("Market Value Over Time")
    plt.xlabel("Year")
    plt.ylabel("Market Value (Million USD)")

    plt.subplot(1, 2, 2)
    plt.plot(df["Year"], df["YoY Growth"], marker='o', linestyle='-')
    plt.title("Year-over-Year Growth Over Time")
    plt.xlabel("Year")
    plt.ylabel("YoY Growth")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    initial_value = float(input("Enter the initial value in million USD: "))
    end_value = float(input("Enter the end value in million USD: "))
    start_year = int(input("Enter the starting year: "))
    end_year = int(input("Enter the ending year: "))

    factors_dict = {}
    num_factors = int(input("Enter the number of custom factors: "))
    for _ in range(num_factors):
        year = int(input("Enter the year for the custom factor: "))
        factor = float(input("Enter the impact factor (1 for no impact): "))
        factors_dict[year] = factor

    result_df = calculate_yearly_values_with_factors(initial_value, end_value, start_year, end_year, factors_dict)
    print(result_df)

    plot_graphs(result_df)
```
